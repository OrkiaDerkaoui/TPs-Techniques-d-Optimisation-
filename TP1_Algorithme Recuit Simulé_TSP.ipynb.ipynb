```python
                    
            #Initialisation : Les paramètres de L'algorithme du Recuit Simulé
N=5
T0= 10
T=T0
Tmin=1e-2
#Tmin=4
tau=1e2
beta=1e-1
k=0
t=0

M= np.array([[0, 120, 220, 150, 210],
            [120, 0, 100, 110, 130],
            [220, 80, 0,  160, 185],
            [150, 0, 160,  0,  190],
            [210, 130, 185, 0,  0]])
             # Calcul de la fonction cout
print(M)

def fonction_cout1():
   global chemin
   energie = 0.0
   xy = np.column_stack((x[chemin],y[chemin]))
   energie = np.sum(np.sqrt(np.sum((xy- np.roll(xy,-1,axis=0))**2,axis=1)))
   return energie

                # Fonction de dépalcement dans le chemin avec Matrice des Distances

def fonction_cout():
   global chemin
   print(chemin)
   energie = 0.0
   c1=0
   for c1 in range(0,N-1) :
        if chemin[c1]!=chemin[c1+1] and M[chemin[c1]][chemin[c1+1]]!=0: 
           print(chemin[c1],chemin[c1+1],M[chemin[c1]][chemin[c1+1]])
           energie=energie+M[chemin[c1]][chemin[c1+1]] 
   energie=energie+M[chemin[c1+1]][chemin[0]]    
   return energie

   
             # Fonction de dépalcement dans le chemin
        
def Deplacement(ville1,ville2):
    print('========================')
    global chemin
    Min = min(ville1,ville2)
    Max = max(ville1,ville2)
    chemin[Min:Max] = chemin[Min:Max].copy()[::-1]
    return
             
```

```python
 # Choix ddes coordonnées des points villes
    
x=np.random.uniform(0, 1, N)
y=np.random.uniform(0, 1, N)
                
    # Choix du chemin initiale
    
chemin = np.arange(N)  # on peut proposer d'autre façon de determiner le chemin intiale
chemin=[2,4,1,3,0]
chemin1 = chemin.copy()
d1=fonction_cout() 

             #Calcul de la distance du chemn initiale

cout_courant = d1
#fonction_cout() 
print('k =', k, 'Chemin Hamiltonien : ',chemin,' Distance du chemin= ', cout_courant, 'Température',T)

                        # Sauvegarde des valeurs pour le traçage
    
Distance = []         # Distance = énergie
Temps = []            # Temps
Temperature = []      # Température

Distance.append(cout_courant)        # Distance = énergie
Temps.append(t)                      # Temps
Temperature.append(T)                # Température
```

```python
ville_depart = chemin[0]            # on fixe la ville de départ
#ville_courante = rd.randint(0,N-1)
decision=''
dE=0
print('k =', 'Chemin Hamiltonien : ',' Distance du chemin= ', 'Température','np.exp(-dE/T)','Décision')
print(k, chemin, cout_courant, T, np.exp(-dE/T),decision)
while  T > Tmin :                
   ville_voisine = rd.randint(1,N-1)  # on peut toujours proposer d'autre techniques ou heuristiques pour choisir le voisin
   ville_courante = rd.randint(1,N-1)
   if (ville_voisine == ville_courante) or ville_voisine == (ville_courante): continue
        
     # Déplacement
   Deplacement(ville_voisine,ville_courante)
     #Calcul de la distance du nouveau
   cout_nouveau = fonction_cout() 
    
      
   if cout_nouveau <= cout_courant: 
     cout_courant= cout_nouveau
   else:  #Critère de Métropolis
    dE = cout_nouveau-cout_courant    
    #beta=rd.uniform(0,1)
    if beta > np.exp(-dE/T): 
      Deplacement(ville_voisine,ville_courante)  # la nouvelle solution (le nouveau chemin) est rejetée.
      decision='Rejet'
    else:    
      cout_courant = cout_nouveau  # On accepte la nouvelle solution (chemin)malgré qu’elle n’améliore pas la fonction  avec la
                                   #probabilité du critère de Métropolis  
      decision='Accept'
  
   # Abaisser la température ======> Refroidissement
   t+=1
   k+=1
   T = T0*np.exp(-t/tau) 
   
   #print('k =', k, 'Chemin Hamiltonien : ',chemin,' Distance du chemin= ', cout_courant, 'Température',T)
   
   print(k, chemin, cout_courant, T, np.exp(-dE/T),decision)
         
   
   Distance.append(cout_courant)        # Distance = énergie
   Temps.append(t)                      # Temps
   Temperature.append(T)                # Température
   
   

#fin de la boucle While

A=[]
B=[]
for k in range(0,N) :
    A.append(chemin1[k])
    B.append(chemin[k])  
A.append(chemin1[0])
B.append(chemin[0])  


                                        #Affichage des résultats finaux

print('                                    Résultat Final                                                  ')
print('Chemin Hamiltonien Initial : ',A,' Distance du chemin Initial = ', d1, 'Température Initiale =',T0)
print('Chemin Hamiltonien Optimal : ',B,' Distance du chemin Optimal= ', cout_courant, 'Température Finale = ',T)
#plt.plot(x[chemin],y[chemin],marker='o',color='g')

plt.figure(1)
plt.subplot(1,2,1)
plt.xticks([])
plt.yticks([])
plt.xlabel('Abscisse x')
plt.ylabel('Ordonné" y')
plt.plot(x[chemin1],y[chemin1],'k')
plt.plot([x[chemin1[-1]], x[chemin1[0]]],[y[chemin1[-1]],y[chemin1[0]]],'k')
plt.plot(x,y,'ro')
plt.title('Chemin Initial)')


plt.subplot(1,2,2)
plt.xticks([])
plt.yticks([])
plt.xlabel('Abscisse x')
plt.ylabel('Ordonné" y')
plt.plot(x[chemin],y[chemin],'g')
plt.plot([x[chemin[-1]], x[chemin[0]]],[y[chemin[-1]],y[chemin[0]]],'g')
plt.plot(x,y,'ro')
plt.title('Chemin Final (Optimisé)')



```
